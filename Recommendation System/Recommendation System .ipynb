```python
import pandas as pd


file_path = r'C:\Users\shady\Downloads\u.data'  
columns = ['USER_ID', 'ITEM_ID', 'RATING', 'TIMESTAMP']

# Since the file is tab-separated, specify the delimiter
df = pd.read_csv(file_path, delimiter='\t', names=columns)

# Add the EVENT_TYPE column with a fixed value 'rating'
df['EVENT_TYPE'] = 'rating'

# Rearrange the columns to match the required format for Amazon Personalize
personalize_df = df[['USER_ID', 'ITEM_ID', 'EVENT_TYPE', 'TIMESTAMP']]

# Save the transformed data to a CSV file 
output_file = 'personalize_interactions.csv'
personalize_df.to_csv(output_file, index=False)

# Display the first few rows of the transformed dataframe
personalize_df.head()
```

```python
import boto3

# AWS credentials should be configured using AWS CLI or environment variables
personalize_runtime = boto3.client('personalize-runtime', region_name="eu-west-1")

# Replace with your actual campaign ARN
campaign_arn = "arn:aws:personalize:eu-west-1:149536478399:campaign/personalized-recommendations"

# List of multiple user IDs from your dataset
user_ids = ["196", "186", "22", "244", "166"]

# Loop through each user and get recommendations
for user_id in user_ids:
    print(f"\n Recommendations for User ID: {user_id}")

    response = personalize_runtime.get_recommendations(
        campaignArn=campaign_arn,
        userId=user_id
    )

    # Print recommended ITEM_IDs
    for item in response['itemList']:
        print(f"  - Recommended Item ID: {item['itemId']}")
```
